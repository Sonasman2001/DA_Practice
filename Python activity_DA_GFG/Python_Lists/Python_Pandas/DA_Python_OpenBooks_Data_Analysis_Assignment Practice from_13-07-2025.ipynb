```python
import os
import pandas as pd
from tqdm.auto import tqdm
```

```python
csv_files = os.listdir('/content/datasets/OpenBook')

main_df_list = []

for csv_file in tqdm(csv_files):

  file_path = 'datasets/OpenBook/' + csv_file

  df = pd.read_csv(file_path)

  main_df_list.append(df)

main_df = pd.concat(main_df_list, ignore_index=True)
```

#### 3.1) Remove Null Values

#### 3.3) Clean Year Column and save Years as a Number

```python
main_df['publish_year'] = main_df['publish_year'].str.replace('First published in ', '').astype(int)

main_df = main_df[main_df['publish_year'] > 1000]
main_df = main_df[main_df['publish_year'] < 2025]
```

```python
main_df['title_id_length'] = main_df['title_id'].str.split('/').str[-1].str.split('?').str[0].str.len()
```

### 4. Data Analysis

#### 4.1) What are the most common book categories in the dataset?

```python
import plotly.graph_objects as go

category_counts = main_df['category'].value_counts().head(10)

fig = go.Figure(go.Bar(
    x=category_counts.values,
    y=category_counts.index,
    orientation='h',
    marker=dict(color=category_counts.values, colorscale='Blues'),
    text=category_counts.values,
    textposition='auto'
))

fig.update_layout(
    title='Top 10 Book Categories',
    xaxis_title='Number of Books',
    yaxis_title='Category',
    font=dict(family='monospace'),
    template='plotly_white',
    hovermode='y unified'
)

fig.show()
```

### üìä Conclusion: Top 10 Book Categories

The bar chart illustrates the distribution of books across the top 10 categories by total number of entries.

Key observations:
- **Juvenile Literature** stands out overwhelmingly with **257,593** books, far exceeding all other categories. This reflects a strong focus on children's and young adult literature in the dataset.
- The next leading categories are **Management** (81,991), **Poetry** (74,624), and **Religion** (71,719), indicating a diverse spread of literary interests.
- **Psychology**, **Music**, and **Finance** are also significantly represented, each with over 65,000 entries.
- **Political Science**, **Mathematics**, and **Cooking** round out the list, with cooking having the lowest among the top 10 at **33,012** books.

Overall, this distribution highlights a major emphasis on educational and creative literary content, with strong representation in both academic and lifestyle-oriented categories.

#### 4.3) How has book publishing evolved over the years?

```python
import plotly.graph_objects as go

publish_year_counts = main_df[main_df['publish_year'] >= 1700]['publish_year'].value_counts().reset_index().sort_values(by = 'publish_year')

fig = go.Figure(go.Scatter(
    x=publish_year_counts['publish_year'],
    y=publish_year_counts['count'],
    mode='lines+markers'
))

fig.update_layout(
    title='Book Publishing Evolution From 1700 Onwards',
    xaxis_title='Publish Year',
    yaxis_title='Number of Books',
    font=dict(family='monospace'),
    template='plotly_white',
    hovermode='x unified'
)

fig.show()
```

### üìà Conclusion: Book Publishing Evolution (1700‚ÄìPresent)

This line chart visualizes the evolution of book publishing from the year 1700 onward.

Key insights:
- **1700‚Äì1900**: The number of published books remained relatively low and steady, indicating limited access to publishing technologies and lower global literacy rates during this period.
- **Post-1900 Growth**: A noticeable surge in publishing activity began around the early 20th century, with steady growth reflecting advancements in printing, education, and global knowledge exchange.
- **Rapid Expansion (1950‚Äì2010)**: The sharpest rise occurred in the second half of the 20th century and early 21st century. This era likely benefited from technological advancements (e.g., computers, internet, self-publishing) and a global push for literacy and education.
- **Peak & Decline**: Publishing appears to have peaked between **2005 and 2015**, reaching over **35,000** books annually. A decline is observed after this peak, particularly sharp around the last few years, which might be due to incomplete recent-year data or changing industry trends (e.g., shift to digital formats not captured in the dataset).

Overall, the data reflects a dynamic and accelerating growth in global knowledge dissemination, with potential future shifts influenced by digital transformation and evolving media consumption habits.


#### 4.4) What percentage of books have never been read, marked as 'Want to read', or 'Currently reading', or 'have read'?

```python
import plotly.graph_objects as go

total_books = len(main_df)
books_with_reviews = len(main_df[main_df['reviews'] > 0])
books_with_no_reviews = len(main_df[main_df['reviews'] == 0])

percentage_with_reviews = (books_with_reviews / total_books) * 100
percentage_with_no_reviews = (books_with_no_reviews / total_books) * 100

labels = ['Books with Reviews', 'Books with No Reviews']
values = [percentage_with_reviews, percentage_with_no_reviews]

# Define colors for the pie chart slices
colors = ['#1f77b4', '#aec7e8'] # Using colors from the 'Blues' palette

fig = go.Figure(data=[go.Pie(labels=labels, values=values, hole=.3, marker=dict(colors=colors))])

fig.update_layout(
    title_text="Percentage of Books with and without Reviews",
    font=dict(family='monospace'),
)

fig.show()
```

### üìù Conclusion: Book Review Coverage

This pie chart illustrates the distribution of books based on the presence of user reviews.

Key observations:
- **Only 21.2%** of books in the dataset have received reviews from readers.
- A significant majority ‚Äî **78.8%** ‚Äî of books have **no reviews** recorded.

üìå **Interpretation**:
- The low review coverage suggests that while many books are being cataloged, only a small fraction engage active reader feedback.
- This may indicate:
  - A need for improved discoverability or promotion of lesser-known books.
  - A gap in community participation, especially for academic or niche titles.
- Increasing user-generated content such as reviews could help other readers make informed choices and boost visibility for more titles.

Encouraging readers to leave feedback can contribute meaningfully to the reading ecosystem and foster a more interactive reading community.

```python
import plotly.graph_objects as go

# Get top authors based on total reading counts
top_authors = main_df.groupby('author')['reviews'].sum().sort_values(ascending=False).head(10).index

# Filter the DataFrame to include only the top authors
top_authors_df = main_df[main_df['author'].isin(top_authors)]

# Group by author and sum the reading stats
author_reading_stats = top_authors_df.groupby('author')[['want_to_read', 'currently_reading', 'have_read']].sum()

fig = go.Figure(data=[
    go.Bar(name='Want to Read', y=author_reading_stats.index, x=author_reading_stats['want_to_read'], orientation='h'),
    go.Bar(name='Currently Reading', y=author_reading_stats.index, x=author_reading_stats['currently_reading'], orientation='h'),
    go.Bar(name='Have Read', y=author_reading_stats.index, x=author_reading_stats['have_read'], orientation='h')
])

fig.update_layout(
    barmode='stack',
    title='Top Authors by Reading Status',
    xaxis_title='Number of Books',
    yaxis_title='Author',
    font=dict(family='monospace'),
    template='plotly_white',
    hovermode='y unified',
    colorway=['#a6cee3', '#1f78b4', '#1f78b4'] # Using colors from the 'Blues' palette
)

fig.show()
```

### üìä Conclusion: Top Authors by Reading Status

The graph above visualizes the top authors based on reader engagement categorized into **"Want to Read"**, **"Currently Reading"**, and **"Have Read"** statuses.

Key insights:
- **Colleen Hoover** leads by a wide margin in total reader interaction, particularly in the *‚ÄúWant to Read‚Äù* category, indicating high interest among readers.
- **Stephen King** and **J.K. Rowling** show a strong balance across all three statuses, reflecting both popularity and high reader commitment to completion.
- **Rick Riordan**, **Roald Dahl**, and **Agatha Christie** have significant ‚ÄúHave Read‚Äù counts, suggesting their works are frequently completed.

- **Robert Greene** and **Robert T. Kiyosaki** show moderate interest with steady numbers across all statuses, consistent with their appeal in nonfiction genres.

This visualization effectively highlights both current reading trends and future reading intentions of users toward popular authors.
